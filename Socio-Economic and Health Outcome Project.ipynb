### *Data Visualisation*

*To under various Socio-econmic factors and its health outcomes we must look at it graphically for wider applications*

Plotting a Histogram to show distribution of Women's Literacy rates over districts

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Women_Literacy'].dropna(), bins=20, color='mediumslateblue')
plt.title("Distribution of Women's Literacy Rate")
plt.xlabel("Women's Literacy (%)")
plt.ylabel("Number of Districts")
plt.show()
```



*   This distribution indicated a bell-shaped curvature which shows that majority of districts have women's literacy rates concentrated in the range of approximately 65% to 85%. There are fewer districts with very low or very high literacy rates, suggesting a significant level of educational development across most districts.

######We shall now look at Scatterplot Diagram to understand relation between Socio-Economic factor and Health Condition

```python
#Scatterplot Relation for Women's Literacy vs Underweight Children
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Women_Literacy',
    y='Children_Underweight',
    data=df,
    color='seagreen',
    alpha=0.7
)
sns.regplot(
    x='Women_Literacy',
    y='Children_Underweight',
    data=df,
    scatter=False,
    color='black'
)
plt.title("Women's Literacy vs. Underweight Children")
plt.xlabel("Women's Literacy (%)")
plt.ylabel("Underweight Children (%)")
plt.show()
```



*  This Scatterplot shows the inverse relation between Women's Literacy and Underweight Children the Trend Line is downward sloping as women's literacy rates rise, the percentage of underweight children generally declines, indicating a potential positive impact of education on child health outcomes. The confidence interval is relatively narrow and consistently follows the trend of the regression line. Which means that the study is accurate


*Graphing a Scatterplot to show relation between Women's Literacy and Anaemia in Children*

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Women_Literacy',
    y='Children_Anaemic',
    data=df,
    color='Blue',
    alpha=0.7
)
sns.regplot(
    x='Women_Literacy',
    y='Children_Anaemic',
    data=df,
    scatter=False,
    color='black'
)
plt.title("Relation of Women's Literacy with Anaemia in Children")
plt.xlabel("Women's Literacy (%)")
plt.ylabel("Anaemic Children (%)")
plt.show()
```



*   This scatter plot shows a negative correlation between women's literacy and the percentage of anemic children. The trend line also slopes downwards as women's literacy increases, the percentage of children suffering from anemia tends to decrease, suggesting that higher maternal education may contribute to better child health in terms of anemia prevention. The confidence interval is also relatively narrow and closely follows the regression line.



*Graphing a Scatterplot to show relation between households with Access to Clean Fuel and Growth Stunt in Children*

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Clean_Fuel_Cooking',
    y='Children_Stunted',
    data=df,
    color='violet',
    alpha=0.7
)
sns.regplot(
    x='Clean_Fuel_Cooking',
    y='Children_Stunted',
    data=df,
    scatter=False,
    color='black'
)
plt.title("Clean Fuel for Cooking")
plt.xlabel("Clean Fuel")
plt.ylabel("Stunting in Children (%)")
plt.show()
```



*   There is a visible negative correlation shown by the downward-sloping black regression line. This suggests that as the proportion of households using clean cooking fuel increases, the percentage of underweight children tends to decrease. The shaded grey area around the line represents the confidence interval for this trend. Which affirms its reliability.


*Graphing a Scatterplot to show relation between households with Access to Clean Fuel and Growth Stunt in Children*

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Electricity_Access',
    y='Children_Stunted',
    data=df,
    color='violet',
    alpha=0.7
)
sns.regplot(
    x='Electricity_Access',
    y='Children_Stunted',
    data=df,
    scatter=False,
    color='black'
)
plt.title("Relation of Households having Electricity with Growth Stunt in Children")
plt.xlabel("Electricity Access (%)")
plt.ylabel("Stunted Growth (%)")
plt.show()
```

######Graphing a Boxplot to combine the district wise data of Socio-Economic Factor and Health Condition among states

Anaemia in Children State-Wise

```python
plt.figure(figsize=(14,7))
major_states = ['Bihar', 'Kerala', 'Jharkhand', 'Tamil Nadu', 'Uttar Pradesh']
tmp = df[df['State/UT'].isin(major_states)]
sns.boxplot(x='State/UT', y='Children_Anaemic', data=tmp, palette="Set3")
plt.title("Children Anaemic (%) by Major State")
plt.xlabel("State")
plt.ylabel("Anaemia in Children (%)")
plt.show()
```



*   This box plot compares the percentage of anemic children across five major states: Bihar, Jharkhand, Kerala, Tamil Nadu, and Uttar Pradesh. Kerala has significantly lower rates of Anaemic Children while Bihar, Jharkhand and Uttar Pradesh shows consistently high rates of Anaemic Children.



Distribution of Stunting in Children by districts with Access to Electricity

```python
df['Electricity_Quartile'] = pd.qcut(df['Electricity_Access'], 4, labels=['Low', 'Mid-low', 'Mid-high', 'High'])

plt.figure(figsize=(8,5))
sns.boxplot(
    x='Electricity_Quartile',
    y='Children_Stunted',
    data=df,
    palette='Pastel1'
)
plt.title("Stunting by Electricity Access Quartile")
plt.xlabel("Electricity Access Quartile")
plt.ylabel("Stunting (%)")
plt.show()
```



*   The box plot compares the percentage of stunting in children across districts grouped by their level of electricity access. Districts with low electricity access have the highest median stunting rates as electricity access increases the median stunting rate consistently decreases. Districts with high electricity access have the lowest median stunting rates.


#####Correlation Matrix Heatmap: Socio-Economic vs Health Outcomes

```python
cols = [
    'Children_Anaemic', 'Children_Stunted', 'Children_Underweight',
    'Women_Literacy', 'Electricity_Access', 'Improved_Sanitation', 'Clean_Fuel_Cooking,'
]
corr = df[cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix: Socio-Economic vs. Health Outcomes")
plt.show()
```

*   This Heatmap shows overall correlation between various Socio-Economic factors and Health Outcomes:
*   Red colors and positive numbers (e.g., 0.71):
Strong positive relationship. As one variable increases,
the other tends to increase.
*   Blue colors and negative numbers (e.g., -0.55): Strong negative relationship. As one variable increases, the other tends to decrease.
*   Values near zero: Weak or no linear relationship.

Women's literacy is an important factor positively linked to a child's health, significantly reducing problems like stunting and anaemia. However, it's not the only solution. Its positive effect works alongside other factors such as improved sanitation and access to clean cooking fuel, which are also crucial for a child's well-being.
